```python
import os
import xml.etree.ElementTree as ET
import copy
```

# This code only finds 1.3 version, because root.get() does not work for v1.0

```python
def extract_xml_files(input_list_pmids, input_list_pmcids, input_dir, output_file):
    CRC_root = ET.Element("CRC_articles")
    previosly_found_pmids = []
    previosly_found_pmcids = []

    for subdir, _, files in os.walk(input_dir):
        for file in files:
            if file.endswith('.xml'):
                xml_file_path = os.path.join(subdir, file)

                try:
                    tree = ET.parse(xml_file_path)
                    root = tree.getroot()

                    # Check the DTD version
                    dtd_version = root.get('dtd-version', 'v1.0')  # Default to version 1.0 if not present
                    print(dtd_version)
                    if dtd_version == 'v1.0':
                        print(dtd_version)
                        root_copy = copy.deepcopy(root)
                        for element in root_copy.iter('article-id'):
                            if ((element.attrib.get('pub-id-type') == 'pmid' and element.text in input_list_pmids and element.text not in previosly_found_pmids) or 
                                (element.attrib.get('pub-id-type') == 'pmc' and element.text in input_list_pmcids and element.text not in previosly_found_pmcids)):  

                                CRC_root.append(root_copy)
                                # Keep track of added pmids and pmcids
                                if element.attrib.get('pub-id-type') == 'pmid':
                                    previosly_found_pmids.append(element.text)
                                if element.attrib.get('pub-id-type') == 'pmc':
                                    previosly_found_pmcids.append(element.text)

                                break  # Break the loop once a valid 'article-id' is found

                    elif dtd_version == '1.3':
                        print(dtd_version)
                        # Extract logic for version 1.3
                        root_copy = copy.deepcopy(root)
                        for element in root_copy.iter('article-id'):
                            if ((element.attrib.get('pub-id-type') == 'pmid' and element.text in input_list_pmids and element.text not in previosly_found_pmids) or 
                                (element.attrib.get('pub-id-type') == 'pmc' and element.text in input_list_pmcids and element.text not in previosly_found_pmcids)):  

                                CRC_root.append(root_copy)
                                # Keep track of added pmids and pmcids
                                if element.attrib.get('pub-id-type') == 'pmid':
                                    previosly_found_pmids.append(element.text)
                                if element.attrib.get('pub-id-type') == 'pmc':
                                    previosly_found_pmcids.append(element.text)

                                break  # Break the loop once a valid 'article-id' is found

                    else:
                        print(f"Unsupported DTD version: {dtd_version} Skipping file: {xml_file_path}")
                        continue

                    # Common logic after DTD version check...
                    CRC_root.append(root_copy)
                    # Keep track of added pmids and pmcids
                    for element in root_copy.iter('article-id'):
                        if element.attrib.get('pub-id-type') == 'pmid':
                            previosly_found_pmids.append(element.text)
                        elif element.attrib.get('pub-id-type') == 'pmc':
                            previosly_found_pmcids.append(element.text)

                except ET.ParseError:
                    print(f"Skipping file {xml_file_path} due to parse error")

    CRC_tree = ET.ElementTree(CRC_root)
    CRC_tree.write(output_file, encoding='utf-8')
```
